# Colorspaces

Let's have a brief introduction into converting to different colorspaces! The video goes into more detail about colorspaces.

Quick Info Link:
https://en.wikipedia.org/wiki/HSL_and_HSV


```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img = cv2.imread('/kaggle/input/picture/dog-puppy-on-garden-royalty-free-image-1586966191.jpg')
```

### Converting to Different Colorspaces

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

**Converting to HSV**
https://en.wikipedia.org/wiki/HSL_and_HSV

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(img)
```

```python
img = cv2.imread('/kaggle/input/picture/dog-puppy-on-garden-royalty-free-image-1586966191.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
plt.imshow(img)
```
